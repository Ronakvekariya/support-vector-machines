```python
import pandas as pd
import numpy as np
import plotly.express as plx
```

```python
data = pd.read_csv('drug200.csv')
```

```python
data
```

```python
data.isna().sum()
```

```python
from sklearn.model_selection  import train_test_split
```

```python
train_in , test_in , train_out , test_out = train_test_split(data[['Age','Sex','BP','Cholesterol','Na_to_K']],data['Drug'] , random_state = 42)
```

```python
categorical =  train_in.select_dtypes('object').columns.tolist()
categorical
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(sparse = False , handle_unknown = 'ignore')
encoder.fit(train_in[categorical])
encoder.categories_
```

```python
cols_names = list(encoder.get_feature_names_out(categorical))
cols_names
```

```python
new_train_in = pd.DataFrame()
new_train_in[cols_names] = encoder.transform(train_in[categorical])
new_train_in
```

```python
new_test_in = pd.DataFrame()
encoder.fit(test_in[categorical])
new_test_in[cols_names] = encoder.transform(test_in[categorical])
new_test_in
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
numeric = train_in.select_dtypes(include = np.number).columns.tolist()
scaler.fit(train_in[numeric])
new_train_in[numeric] = scaler.transform(train_in[numeric])
```

```python
new_test_in[numeric] = scaler.fit_transform(test_in[numeric])
new_test_in
```

```python
new_train_in
```

```python
new_train_in.describe()
```

```python
len(numeric)
new_train_in
encoder.categories_
```

```python
from sklearn.feature_selection import chi2

# Prepare data (X: features, y: target variable)
X = new_train_in[cols_names]
y = train_out 
# Compute Chi-square statistics and p-values
chi2_stats, p_values = chi2(X, y)


# Get indices of top K features based on chi2 statistics
top_k_features = chi2_stats.argsort()[-3:]

# Print selected features
print(f"Top {4} features based on Chi-square:")
for i in top_k_features:
    print(f"- Feature {i}" , train_in.columns[i])

```

```python
from sklearn.datasets import load_iris
from sklearn.feature_selection import f_classif

# Load the iris dataset

X = new_train_in[numeric]
y = train_out

# Compute the ANOVA F-value and p-value for each feature
f_values, p_values = f_classif(X, y)

# Print the F-value and p-value for each feature
for i in range(X.shape[1]):
    print("Feature %d: F-value=%.2f, p-value=%.2f" % (i, f_values[i], p_values[i]))
```
